#### Paso 1: Instalación de paquetes necesarios


```julia
using DelimitedFiles
using Plots

# Carga de datos desde un archivo delimitado por tabulaciones
data_tofit = DelimitedFiles.readdlm("Week3_PR_Data.dat", '\t')
```

```julia
# Asigna la primera columna de data_tofit a la variable valX.
valX = data_tofit[:, 1]

# Itera sobre el número de elementos en valX (que es igual al número de filas en data_tofit).
for i = 1:length(valX)
    # Imprime el valor de la primera columna seguido del valor de la segunda columna para cada fila de data_tofit.
    println(data_tofit[i, 1], " ", data_tofit[i, 2])
end
```

```julia
# Asigna la primera columna de los datos de data_tofit a la variable x.
x = data_tofit[:, 1]

# Asigna la segunda columna de los datos de data_tofit a la variable y.
y = data_tofit[:, 2]
```

```julia
# Genera un gráfico de dispersión para los datos x e y.
# Cada punto del gráfico representa una observación (x[i], y[i]).
Plots.scatter(x, y)
```

```julia
function parabfit(x)
    """
    Calcula el valor de una función cuadrática (parábola) para un valor dado de x.

    Parámetros:
    - x: Valor de la variable independiente.

    Retorna:
    - Valor calculado de la función cuadrática con coeficientes a, b, y c.

    Se asumen los coeficientes a, b, y c están previamente definidos.
    """
    return a*x^2 + b*x + c
end
```

```julia
# Definición de coeficientes
a, b, c = 1, 1, 1

# Gráfico de la función cuadrática (parábola)
plot(parabfit, -5, 5, 
     title="Gráfico de la función cuadrática",
     xlabel="x",
     ylabel="y",
     label="y(x) = ax^2 + bx + c")
```

```julia
# Gráfico de dispersión de los datos originales
scatter(x, y, label="Datos")

# Asignación de coeficientes para la función parabólica
a = -1
b = 3.9
c = -2.3

# Gráfico de la función parabólica ajustada junto con los datos originales
plot!(parabfit, 0, 6,
      label="Ajuste parabólico",
      xlabel="x",
      ylabel="y")
```
