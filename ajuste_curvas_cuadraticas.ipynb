#### Importación de paquetes y datos


```julia
using Plots
gr()  # Activar el backend GR para usar con Plots
using DelimitedFiles

data_tofit = readdlm("Week3_PR_Data.dat", '\t')
rows, cols = size(data_tofit)
```

#### Visualización de datos

```julia
for r = 1:rows
    println(data_tofit[r, :])
end
```

```julia
for r = 1:rows
    println(data_tofit[r,:])
end
```

#### Preparación de datos

```julia
x,y = data_tofit[:,1], data_tofit[:,2]
```

#### Gráfico de datos dispersos


```julia
scatter(x, y, labels="Datos a ajustar", xlabel="x", ylabel="y", title="Datos a ajustar con ajuste parabólico")
```

#### Definición de la función de ajuste parabólico


```julia
function parabfit(x)
    return a*x^2+b*x+c
end
```

#### Ajuste inicial de la curva

```julia
a = b = c = 1
plot(parabfit, -5, 5, labels="y(x) = x^2 + x + 1")
```

#### Iteración y ajuste de parámetros

```julia
# Probamos diferentes valores para a, b y c
# a, b, c = -1, 1.5, 0
# a, b, c = -1, 2.5, 0
# a, b, c = -1, 3, 0
# a, b, c = -1, 3, -1
# a, b, c = -0.8, 3, -1.5
a, b, c = -0.8, 3, -1.75

scatter(x, y, labels="Datos a ajustar")
plot!(parabfit, -2, 6, labels="Curva ajustada")
```

#### Conclusiones y comentarios
Se muestra el proceso de ajuste de la curva parabólica a los datos, explicando cómo se seleccionaron y ajustaron los parámetros a, b y c para obtener la mejor aproximación visual de la curva a los datos.
